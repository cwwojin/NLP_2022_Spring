```python
import os
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords

nltk.download('stopwords')
tokenizer = TreebankWordTokenizer()
```

```python
cwd = os.getcwd()
print(cwd)
dir_bloomberg = "20061020_20131126_bloomberg_news"
ls = sorted(os.listdir(dir_bloomberg))
files = []
word_occurences_uncased = []
stop_words = set(stopwords.words('english'))
#stop_words_punct = set(stopwords.words('english') + list(string.punctuation))
re_punkt = re.compile("[" + string.punctuation + "]+")

#corpus_name = "bloomberg_corpus_uncased.txt"
corpus_name = "bloomberg_corpus_v2.txt"

if os.path.exists(corpus_name):
    os.remove(corpus_name)

with open(corpus_name, "w+", encoding="UTF-8") as corpus_file :
    for d in ls :
        print(d)
        dir_date = dir_bloomberg + "/" + d
        files = os.listdir(dir_date)
        for fname in files :
            with open(dir_date + "/" + fname, 'r', encoding="UTF-8") as f :
                contents = f.readlines()
                if len(contents) <= 4 : continue
                article = re.sub(r"\n", " ", "".join(contents[4:]))
                match = re.search("To contact the", article)
                if match is not None :
                    article = article[:match.start()]
                article_tokenized = TreebankWordTokenizer().tokenize(article)
                article_uncased = [i.lower() for i in article_tokenized if (i.lower() not in stop_words) and (not re_punkt.fullmatch(i))]
                for word in article_uncased :
                    #print(word, end=" ") 
                    corpus_file.write(word + " ")
                corpus_file.write("\n")
```
