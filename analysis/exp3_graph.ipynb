```python
import matplotlib.pyplot as plt
```

```python
micro = [0.798,
0.802,
0.804,
0.796,
0.792
]
micro_diff = [mic-0.708 for mic in micro]

weighted = [0.794,
0.791,
0.796,
0.790,
0.797
]
weighted_diff = [wei-0.708 for wei in weighted]
size = range(0,125, 25)
y = range(70,110,10)
y_size = [y_val/1000 for y_val in y]
```

```python

plt.plot(size, micro_diff, 'bo-', label = 'Micro F1 Diff')
plt.plot(size, weighted_diff, 'ro-', label = 'Weighted F1 Diff')
# plt.axhline(0.0, 0, 1, color='gray', linestyle='--', linewidth=2)
plt.xlabel('Ratio between DADA/TF-IDF')
plt.ylabel('Performance gain')
plt.xticks(size)
plt.yticks(y_size)
plt.grid(True)
plt.legend()
plt.show()
```
