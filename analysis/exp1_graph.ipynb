```python
import matplotlib.pyplot as plt
```

```python
micro_diff = [0.148,
0.098,
0,
0.023,
-0.016,
0.011,
-0.005,
-0.005
]

weighted_diff = [0.277,
0.141,
-0.003,
0.026,
-0.016,
0.014,
-0.01,
-0.01,
]
size = range(500,4500,500)
```

```python

plt.plot(size, micro_diff, 'bo-', label = 'Micro F1 Diff')
plt.plot(size, weighted_diff, 'ro-', label = 'Weighted F1 Diff')
plt.axhline(0.0, 0, 1, color='gray', linestyle='--', linewidth=2)
plt.xlabel('Size of labeled dataset')
plt.ylabel('Performance gain')
plt.grid(True)
plt.legend()
plt.show()
```
