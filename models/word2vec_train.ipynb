```python
import os

corpus_sents = []
corpus_name = "bloomberg_corpus_v2.txt"

with open(corpus_name, "r", encoding="UTF-8") as f :
    while True:
        line = f.readline()
        if not line: break
        corpus_sents.append(line.split(" "))
```

```python
from gensim.test.utils import common_texts
from gensim.models import Word2Vec

model = Word2Vec(sentences=corpus_sents)
model.wv.save_word2vec_format("word2vec.bin", binary=True)
```
